```python
sql = "select * from wiki"
cursor.execute(sql)
```

```python
result_wiki = cursor.fetchall()
```

```python
result_wiki[0]
```

```python
import nltk
# nltk.download('punkt')
tokens = nltk.word_tokenize(result_wiki[0][2])
tokens
```

```python
print(type(tokens[1]))
print(tokens[1])
print(type(result_wiki[0][0]))
print(result_wiki[0][0])
```

```python
fdist = nltk.FreqDist(tokens)
fdist
```

```python
print(len(tokens))
print(len(fdist))
```

```python
test_tokens = nltk.word_tokenize("I am a iron man and good man and I am a kind man")
test_tokens_dictionary = nltk.FreqDist(test_tokens)
test_tokens_dictionary
```

```python
print("(term,freq)")
len(test_tokens_dictionary)
for term, freq in test_tokens_dictionary.items():
    print(term, end='')
    print(",",freq)    
```
